```python
import os
from PIL import Image
import random
```

```python
files = [i for i in os.listdir(r"Ground Truth")]
```

```python
for file in files:
    print(file)
    im = Image.open(r"Ground Truth\%s"%file)
    pixels = im.load()
    for i in range(im.size[0]): # for every pixel:
        for j in range(im.size[1]):
            if pixels[i,j][0] < 190 or pixels[i,j][1] > 40 or pixels[i,j][2] >50:
                pixels[i,j] = (0,0,0)
            else:
                pixels[i,j] = (255,0,0)
        im.save(file, 'JPEG')
```

```python
random.shuffle(files)
split = int(len(files)*.8)
train_data = files[:split]
val_data = files[split:]

train = open("train.txt",'w')
train.write('\n'.join(train_data))
train.close()

val = open("val.txt",'w')
val.write("\n".join(val_data))
val.close()

train_data.extend(val_data)

trainval = open("trainval.txt",'w')
trainval.write('\n'.join(train_data))
trainval.close()
```
